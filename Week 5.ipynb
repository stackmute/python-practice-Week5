```python
import glob
import pandas as pd

file_list = glob.glob("data/sales_*.csv")
dfs = [pd.read_csv(file) for file in file_list]

all_data = pd.concat(dfs, ignore_index=True)
all_data["date"] = pd.to_datetime(all_data["date"])
```

```python
summary = all_data.groupby("product")["sales"].agg(["sum", "mean"])
summary
```

```python
summary.to_csv("result_sales_summary.csv")
```

## Output
- The aggregated sales results were saved as `result_sales_summary.csv`.
- This file can be reused for reporting or further analysis.
